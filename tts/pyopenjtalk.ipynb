```python
import pyopenjtalk
from scipy.io import wavfile
import numpy as np
```

```python
x, sr = pyopenjtalk.tts("下長ね沙羅さん、本日はご結婚おめでとうございます。村田ちょーざとこれからもよろしくお願いします。本日は新宿区役所にご来場いただき誠にありがとうございます。婚姻届の窓口は向かって左になります。どうぞよろしくお願いします。", run_marine=True)
out_f = "test.wav"
wavfile.write(out_f, sr, x.astype(np.int16))
```

```python
from IPython.display import Audio
Audio(out_f, autoplay=True)  # dose not play...
```
